```python
import pandas as pd
import plotly.express as px
import pandas as pd
import plotly.io as pio
import dash
from jupyter_dash import JupyterDash
from dash import dcc 
from dash import html
from dash.dependencies import Input, Output
```

## Loading Data

```python
df = pd.read_csv('UdemyCoursesData.csv')
df.head().T
```

```python
df.info()
```

## Data Pre-processing

```python
df.drop(columns=['url',
                 'primary_category_link',
                 'primary_subcategory_link',
                 'instructor_Link',
                 'what_you_will_learn',
                 'description_non_formatted',
                 'reviews',
                 'small_image_link',
                 'medium_image'
                ], inplace=True)

df.dropna(inplace=True)

df["course_creation_date"]=  pd.to_datetime(df["course_creation_date"])

df['year']= df.course_creation_date.dt.year

df['month'] = df.course_creation_date.dt.month

df['count']=1

df['is_paid']='free'
x= lambda u : "paid" if  u != 0 else 'free'
df['is_paid']=df['price'].map(x)


df.head().T
```

```python
df.info()
```

```python
def get_options(lst):
    dict_list = []
    for i in lst:
        dict_list.append({'label': i, 'value': i})

    return dict_list
```

```python
app = dash.Dash(external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])
app.layout = html.Div([
    
     html.Div(style={
                         "height": "60px",
                         "width": "auto",
                         "margin-bottom": "25px",
                     },
                     ),
    
    
     html.Div([
        html.Img(src='https://cdn.freebiesupply.com/logos/large/2x/udemy-2-logo-black-and-white.png' ,
                 style={'width':'200px','margin-bottom':'50px'})
    ],style={'textAlign':'center'}),


    html.Div(dcc.Slider(
        id='year--slider',
        min=2011,
        max=2020,
        value=2019,
        marks={str(year): str(year) for year in df['year'].unique()},
        step=None,
        included=False
        
            
    ),style={'padding-left':'30%','padding-right': '25%','width':600}),
    
    
    html.Div([   
         
    html.Div([
        
#         Courses per year        
        
        html.Div([
        
            html.P("Courses"),
            
            html.Br(),
            
            html.P(id='courses',style={'font-size':'20px'})
            
        
        ],style={
            'height': 150,
            'textAlign':'center',
            'margin-left':'4%',
            'background-color':'purple',
            'padding': 30 ,
            'color':'white',
            'font-size': '12px',
            'font-weight' : 'bold',
            'border-radius': '20px'
        },className='three columns'),
        
        
#         Subscribers per year 
        
         html.Div([
        
            html.P("Subscribers"),
            
            html.Br(),
            
            html.P(id='subscribers',style={'font-size':'20px'})
            
        
        ],style={
            'height': 150,
            'margin-left':'20px',
            'textAlign':'center',
            'background-color':'#AE2012',
            'padding': 30 ,
            'color':'white',
            'font-size': '12px',
            'font-weight' : 'bold',
             'border-radius': '20px'
        },className='three columns'),
        
        
        
        
#         Average Rating per year
        
        
         html.Div([
        
            html.P("Average Rating"),
            
            html.Br(),
            
            html.P(id='rating',style={'font-size':'20px'})
            
        
        ],style={
            'height': 150,
            'margin-left':'20px',
            'textAlign':'center',
            'background-color':'orange',
            'padding': 30 ,
            'color':'white',
            'font-size': '12px',
            'font-weight' : 'bold',
             'border-radius': '20px'
        },className='three columns'),
        
        
        
        
#         Average course price per year
        
        
         html.Div([
        
            html.P("Average price"),
            
            html.Br(),
            
            html.P(id='price',style={'font-size':'20px'})
            
        
        ],style={
            'height': 150,
            'margin-left':'20px',
            'textAlign':'center',
            'background-color':'blue',
            'padding': 30 ,
            'color':'white',
            'font-size': '12px',
            'font-weight' : 'bold',
             'border-radius': '20px'
        },className='three columns'),
        
    ],className='container'),
        
    ],style={"margin-top":35,"margin-bottom": 50,'textAlign':'center'}),
        
#                 

#            
    
    
#     Two Graphes
    
    
html.Div([
        
        
    html.Div([
        
        
        html.Div(dcc.Graph(id='level'))],className='six columns',
        style={'height': 450,'margin-left':0}),
        
        html.Div([html.Div(dcc.Graph(id='category'))],className='six columns',
                 style={'height': 450,}),],style={"width":"90%","margin": "auto","font-size": "12px"}),
    
    
## dropdown
html.Div([
        html.Div([
                 dcc.Dropdown(id='catselector',options=get_options(df["primary_category"].unique()),
                              value='development',style={'textAlign':'center', 'margin-top': 50 , 'margin-bottom': 50}),

            ]),],style={"width":"90%",'margin':"auto" , "padding-top": "50px"}),

    
     html.Div([
        
                html.Div([

                    dcc.Graph(id='sunbrst' ,style={'height': 450})

                ],className='six columns',style={'height': 450}),


                html.Div([

                    dcc.Graph(id='catyear',style={'height': 450})

                ],className='six columns',style={'height': 450}),
        
        
#     html.Div(dcc.Graph(id='sunbrst'),className='four columns',style={'margin-left':0})  
    ],className="container",style={"width":"90% !important","margin": "auto","font-size": "12px"}),
    
   
],


style={'background-color':'black' , 'margin': '-10px'})
```

```python
@app.callback(Output('courses', 'children'),
              Input('year--slider', 'value'))
def courses(selected_year):
    number_ofcourses = len(df[df.year == selected_year])
    
    return number_ofcourses


@app.callback(Output('subscribers', 'children'),
              Input('year--slider', 'value'))
def subscribers(selected_year):
    number_of_subscribers = df[df.year == selected_year].num_subscribers.sum()
    
    return number_of_subscribers


@app.callback(Output('rating', 'children'),
              Input('year--slider', 'value'))
def rating(selected_year):
    avgrate=df[df.year == selected_year].avg_rating.mean()                
    
    return "{:.2f}".format(avgrate)


@app.callback(Output('price', 'children'),
              Input('year--slider', 'value'))
def price(selected_year):
    price_avg = df[df.year == selected_year].price.mean()
    
    return "{:.2f}".format(price_avg)
```

```python
# ['ggplot2', 'seaborn', 'simple_white', 'plotly','plotly_white', 'plotly_dark', 'presentation', 'xgridoff','ygridoff', 'gridon', 'none']
temp="plotly_dark"
colors=['#113CFC', '#ffc300', '#53555B', '#23262E', '#2d3038', '#19D3F3', '#CA6702', '#BB3E03', '#AE2012', '#9B2226']
########################################################################################
@app.callback(Output('level', 'figure'),
              Input('year--slider', 'value'))
def update_figure(selected_year):
    filtered_df = df[df.year == selected_year].groupby(['course_level',"is_paid"],as_index=False).sum().sort_values(by=['num_subscribers'])

    fig=px.bar(filtered_df,x='course_level',y='num_subscribers',color='is_paid',barmode='group',hover_name="num_subscribers",
              labels={
                     "course_level": "Course Level",
                     "num_subscribers": "Number of Subscribers",
                     "is_paid": "Price"
                 },
                title="Number of subscribers per course level" ,
               template=temp,
               color_discrete_sequence=colors
              )
    fig.update_layout(autosize=False)
    return fig
########################################################################################
@app.callback(Output('category', 'figure'),
              Input('year--slider', 'value'))

def update_figure2(selected_year):
    filtered_df = df[df.year == selected_year].groupby(['primary_category','is_paid',"year"],as_index=False).sum().sort_values(by=['num_subscribers'] ,ascending=False)
    fig=px.bar(filtered_df,y='primary_category',x='num_subscribers',color='is_paid',barmode='group',
               labels={
                     "primary_category": "Course Primary Category",
                     "num_subscribers": "Number of Subscribers",
                     "is_paid": "Price"
                 },
                title="Subscribers per category" 
              ,template=temp,
              color_discrete_sequence=colors)
    fig.update_layout(autosize=False)
#     fig.update_layout(transition_duration=500)

    return fig
########################################################################################
@app.callback(Output('sunbrst', 'figure'),
              Input('catselector', 'value'))

def update_figure3(selected_category):
    filtered_df = df[df.primary_category==selected_category].groupby(['primary_category','primary_subcategory'],as_index=False).sum()
    fig=px.sunburst(filtered_df,path=["primary_category",'primary_subcategory'],
               labels={
                     "primary_category": "Course Primary Category",
                     "num_subscribers": "Number Of Subscribers ",
                     "is_paid": "Price"
                 },
                title=" Categories vs. Sub-categories"
                ,template=temp,
                   color_discrete_sequence=colors)

#     fig.update_layout(transition_duration=500)
    fig.update_layout(autosize=False,)
    return fig



########################################################################################
@app.callback(Output('catyear', 'figure'),
              Input('catselector', 'value'))

def update_figure4(selected_category):
    
    filtered_df=df[df.primary_category==selected_category].groupby(['year','primary_category',"is_paid"],as_index=False).sum()
    fig=px.line(filtered_df,x='year',y='count',markers=True,hover_name="primary_category",color="is_paid",
               labels={
                     "year": "Year",
                     "count": "Number of Courses",
                     "is_paid": "Price"
                 },
                title="Number of courses per year for each category"
                ,template=temp
               ,color_discrete_sequence=colors)


    fig.update_layout(autosize=False,)
    return fig


```
